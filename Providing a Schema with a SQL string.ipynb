As you already know, CSV data is stored as text. So if you don't select the option to infer the schema, then everything gets read in as strings in the code cells as shown in the above cell.
I'm reading in the country's population CSV file from the storage volume, and here it is displayed as a dataframe.

As you can see, all of the columns have been read in as strings.
Inferring the schema is an easy way to assign more appropriate data types, but it involves additional reads of the data, and it might not always assign the data types that are the most suitable.

It's always advisable to specify the desired “schema” on read method of spark while reading the file
To do that, you can use the “schema” method and provide an SQL DDL string.
DDL stands for Data Definition Language.

The string you provide simply needs to contain the columns and their data types in the order that they're
being read in.


```python
df = spark.read.schema("country_id int, name string, nationality string, country_code string, iso_alpha2 string, capital string, population int, area_km2 int, region_id int, sub_region_id int").load(path, format="csv", header="true")
```

```python
#Another method is to define schema a variable and pass that variable into the schema method.
Schema = "country_id int, name string, nationality string, country_code string, iso_alpha2 string, capital string, population int, area_km2 int, region_id int, sub_region_id int"

df1 = spark.read.schema(Schema).load(path, format="csv", header="true")                                  
```

Using Python object - Another method is to define schema a variable and pass that variable into the schema method.

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

#Another method is to define schema a variable and pass that variable into the schema method.

Schema = StructType(
                    [ 
                        StructField("country_id", IntegerType(), False), 
                        StructField("name", StringType(), True), 
                        StructField("nationality", StringType(), True), 
                        StructField("country_code", StringType(), True), 
                        StructField("iso_alpha2", StringType(), True), 
                        StructField("capital", StringType(), True), 
                        StructField("population", IntegerType(), True), 
                        StructField("area_km2", IntegerType(), True), 
                        StructField("region_id", IntegerType(), True), 
                        StructField("sub_region_id", IntegerType(), True)
                    ]
        )
```

```python
df = spark.read.schema(Schema).load(path, format="csv", header="true")
```
