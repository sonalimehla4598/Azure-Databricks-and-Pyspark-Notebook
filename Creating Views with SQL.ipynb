```python
#Syntax
%sql
CREATE VIEW <catalog-name>,<schema-name>.<view-name> AS
SELECT query
```

```python

%sql
CREATE VIEW population_metrics1.default.top_10_population AS
SELECT * FROM population_metrics1.default.countries_population ORDER BY population DESC LIMIT 10
```

```python
%sql
SELECT * FROM population_metrics1.default.top_10_population
```
