```python
%sql 
-- Dropping a Catalog
-- Restrict/Cascade - means that by default choice is Restrict. WHich means that if there are any objects in the catalog, then the drop will fail. If you want to drop the catalog and all the objects in it, then you can use the CASCADE option.
-- Restrict can delete only empty catalog
-- CASCADE can delete non-empty catalog

DROP CATALOG [IF EXISTS] <catalog-name> [RESTRICT | CASCADE];
```

```python
%sql
DROP CATALOG test_catalog CASCADE
```

```python
%sql 
-- Dropping a Schema 

DROP SCHEMA [IF EXISTS] <catalog-name> [RESTRICT | CASCADE];
```

```python
%sql 
-- Example to drop Schema 

DROP SCHEMA test_catalog.test_schema
```

```python
%sql 
-- Dropping a Volume 

DROP VOLUME [IF EXISTS] <volume-name>
```

```python
%sql 
-- Dropping a View 

DROP VIEW [IF EXISTS] <view-name>
```
