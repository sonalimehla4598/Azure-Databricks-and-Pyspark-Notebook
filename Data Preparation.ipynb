Creating the country_regions table in Unity Cataog

```python
#Import Modules
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

Creating country_regions table in Unity Catalog

```python
path = "/Volumes/population_metrics1/landing/datasets/countries_dataset/csv_data/country_regions/country_regions.csv"
Schema = StructType(
  [
    StructField("Id", IntegerType(), True),
    StructField("name", StringType(), True)
  ]
)

df = spark.read.load(path=path,schema=Schema, format="csv", header=True)
df.display()
```

```python
df.write.saveAsTable("population_metrics1.default.country_regions")
```

Create country_sub_regions table in Unity Catalog

```python
path = "/Volumes/population_metrics1/landing/datasets/countries_dataset/csv_data/country_sub_regions/country_sub_regions.csv"

Schema = StructType(
    [
        StructField("Id", IntegerType(), True),
        StructField("name", StringType(), True)
    ]
)

df = spark.read.load(path=path,schema=Schema, format="csv", header=True)

df.write.saveAsTable("population_metrics1.default.country_sub_regions")
```
