```python
#Read the table
df = spark.read.table("population_metrics1.default.countries_population")
```

```python
#Select few columns from the dataframe 
df.select("country_id","name","population","area_km2").display()
```

We can also perform the basic tranformations and operation on these column by providing an expression in the select() method.

```python
from pyspark.sql.functions import col
df_selected = df.select(col("country_id"),
                        col("name"),
                        col("population"),
                        col("area_km2"),
                        (col("population")/col("area_km2")).alias("population_density"))
df_selected.display()
```

```python
from pyspark.sql.functions import col, upper
df_selected = df.select(col("country_id"),
                        upper(col("name")).alias("country_name"),
                        col("population"),
                        col("area_km2"),
                        (col("population")/col("area_km2")).alias("population_density"))
df_selected.display()
```

```python
df.selectExpr(
    "country_id",
    "upper(name) as country_name",
    "population",
    "area_km2",
    "(population/area_km2) as population_density"
).display()
```
