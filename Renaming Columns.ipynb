```python
#So here in this block I have used a mixture in the select() of using DataFrame attribute access and string. We can do mix and match

df = spark.read.table("population_metrics1.default.countries_population")
df.select(df.country_id.alias("id"), df.name.alias("country_name"), "population","area_km2").display()
```

Aliasing of column doesnot impact the source data. The column name changes only get persisted when you write the Dataframe to storage. 

```python
#Another way to alias the column names is to use the withColumnRenamed() method. which is part of DataFrame API, we can use this method on DataFrame object.

df.withColumnRenamed("country_id","id").withColumnRenamed("name","country_name").display()
```

We can also use withColumnsRenamed() instead of withColumnRenamed() to rename multiple columns in one go.

```python
df.withColumnsRenamed(
  {
    "country_id":"id",
    "name":"country_name"
  }
).display()
```
