```python
%sql
-- Syntax for Creating Catalog

CREATE CATALOG [ IF NOT EXISTS ] <catalog-name>
    [MANAGED LOCATION '<location-path>']
    [COMMENT <comment>];
```

```python
%sql
-- Create an empty catalog

CREATE CATALOG test_catalog
MANAGED LOCATION 'abfss://unity-catalog-storage@dbstorage22xcfpki3fok6.dfs.core.windows.net/7405619920059625'
```

```python
%sql 
-- Syntax for Creating Schema
CREATE { DATABASE | SCHEMA } {IF NOT EXISTS} <catalog-name>.<schema-name>
    [MANAGED LOCATION '<location-path>' | LOCATION '<location-path>]
    [COMMENT <comment>]
    [WITH DBPROPERTIES ( <property-key = property_value [ , ... ]> )];
```

```python
%sql 
--Create a schema inside newly created test catalog

CREATE SCHEMA test_catalog.test_schema
```

```python
%sql 
-- Syntax for Creating Volume

CREATE VOLUME <catalog>.<schema>.<volume-name>
LOCATION '<location-path>'
```

```python
%sql 
--Create volume in test_Schema 

CREATE VOLUME test_catalog.test_schema.test_volume
```
