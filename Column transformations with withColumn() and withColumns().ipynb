```python
df = spark.read.table("population_metrics1.default.countries_population")
```

Let suppose we need to transform the name column, and wanted to display values in uppercase

```python
df.display()
```

```python
#We need to import the upper function because it is the part of functions API
from pyspark.sql.functions import upper
df.withColumn("name", upper("name")).display()
```

```python
df.withColumn("country_name", upper("name")).display()
```

```python
df.withColumn("country_name", upper("name")).select("country_name","population","area_km2").display()
```

```python
df.withColumn(
    {
         "country_name": upper("name"),
         "population_density": col('population')/col('area_km2')
    }
)
```

```python
#Now let suppose if I want to create a new column population_density which is the result of population/area_km2
#We use col function when we need to perform operations between two columns. Like here to calculate population_density, we need to divide population and area_km2

from pyspark.sql.functions import col
df.withColumn("country_name", upper("name")).\
    withColumn("population_density",col("population")/col("area_km2")).\
        select("country_name","population","area_km2","population_density").display()
```

```python
#Alteranate way to write the above code by creating multiple dataframes and assigning the result of one statement to single dataframe and referring that dataframe in further statements.

df_1 = df.withColumn("country_name", upper("name"))
df_2 = df_1.withColumn("population_density",col("population")/col("area_km2"))
df_3 = df_2.select("country_name","population","area_km2","population_density")
df_3 = display()

```

So above code operation results in creating four dataframes. Instead we could have just used a single select(). Let us show in below cell how we can achieve the same result using single select().


```python
#This approach is more efficient than using multiple withColumn() combined with a select() method.

df.select(
  df.name.alias("country_name"),
  df.population,
  df.area_km2,
  df.population / df.area_km2.alias("population_density")).display()
```
