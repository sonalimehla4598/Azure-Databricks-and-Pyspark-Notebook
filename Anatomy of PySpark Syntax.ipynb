### 📌 Step by Step Approach

The step-by-step approach breaks your code into clear, discrete stages—first ingest, then transform, then persist—by assigning each intermediate result to a named variable. 

That modular style makes it easy to inspect and debug at each point.

```python
#As we know that spark object, a spark session is automatically available in your notebook. This spark session gives us the access to the core operations such as reading, writing as well as Dataframe operations. 
#Here I am using the read method which is available as part of the spark session. This method allows me to read in data as a dataframe. I specified the format as csv and then the path to the file. This will return a Dataframe object which I can perform additional methods and functions on. 
# We have assigned the resulting dataframe to a variable called df.
df = spark.read.format("csv").load("/path/to/file.csv")
```

```python
#Here we are performing a simple filter operation on the same object df.
df = df.filter("Age > 30")
```

```python
#And finally writing the resulting dataframe into a storage location specified here in csv format.
df.write.format("csv").save("/path/to/directory/")
```

In this example above I'm doing everything step by step, so this is the step by step approach. Everything is done in stages.
The first line reads the CSV file into a DataFrame object.
The second line filters the record, and then the 
final line writes the resulting dataframe as CSV into a storage location path.

### 📌 Chaining Methods
Method chaining is a “fluent” interface pattern where each method returns an object—often the same instance or a new immutable one—so you can immediately call the next operation on it. 

Instead of breaking your logic into named, intermediate variables, you stitch calls together in a linear flow that reads like a pipeline: each step consumes the output of the previous one. 

This style makes the overall sequence of transformations concise and expressive, emphasizes the order of operations, and reduces boilerplate—though for very complex or conditional logic you might still choose to pull out intermediate results for clarity or debugging.

```python
spark.read.format("csv").load("/path/to/file.csv").filter("Age > 30").write.format("csv").save("/path/to/directory/")
```

You can use backslashes to break a long method chain into readable, indented lines.

In Python, a backslash (\) indicates that a statement continues onto the next line. Without it, Python would assume the code ends at the newline and you’d get a syntax error.

```python
spark.read.format("csv").load("/path/to/file.csv").\
filter("Age > 30").\
write.format("csv").save("/path/to/directory/")
```
